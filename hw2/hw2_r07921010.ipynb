```python
# problem3
amount = 0
volume = 0
am_ratio = 0
vol_ratio = 0
col1 = '最高價'
col2 = '成交股數'
for i in range(1,len(df)):
    abs_amount = np.abs(df[col1][i]-df[col1][i-1])
    abs_volume = np.abs(df[col2][i]-df[col2][i-1])
    rt_amount = df[col1][i] / df[col1][i-1]
    rt_volume = df[col2][i] / df[col2][i-1]
    if abs_amount > amount:
        amount = abs_amount
        index = i
    if abs_volume > volume:
        volume = abs_volume
        ind_vol = i
    if rt_amount > am_ratio:
        am_ratio = rt_amount
        rt_amt_idx = i
    if rt_volume > vol_ratio:
        vol_ratio = rt_volume
        rt_vol_idx = i
print(amount, df['日期'][index])
print(volume, df['日期'][ind_vol])
print(am_ratio, df['日期'][rt_amt_idx])
print(vol_ratio, df['日期'][rt_vol_idx])
```

```python
# problem 3
diff_highest_price = df['最高價'].diff().abs().max()
diff_highest_price_date = df['最高價'].diff().abs().idxmax()
diff_daily_trade = df['成交股數'].diff().abs().max()
diff_daily_trade_date = df['成交股數'].astype('float64').diff().abs().idxmax()
print(diff_highest_price, df['日期'][diff_highest_price_date])
print(diff_daily_trade, df['日期'][diff_daily_trade_date])
```

```python
# problem 4 [成交股數] [關盤-開盤價] [最高-最低]
df_oc = [abs(df['收盤價'][i] - df['開盤價'][i]) for i in range(len(df))]
df_maxmin = [abs(df['最高價'][i] - df['最低價'][i]) for i in range(len(df))]
df['盤差'] = df_oc
df['高低差'] = df_maxmin
mn_vol = df['成交股數'].mean()
mn_oc = df['盤差'].mean()
mn_maxmin= df['高低差'].mean()
result = []
df2 = df[['成交股數','盤差','高低差']]
df2_normal = (df2 - df2.mean())/ (df2.std())
df2_normal
for i in range(len(df)):
    res = (df2_normal['成交股數'][i]**2 + (df2_normal["盤差"][i])**2 + (df2_normal['高低差'][i])**2)**0.5
    result.append((i,res))
result
sort_result = sorted(result,key=lambda x:x[1],reverse=True)
print(sort_result[:5])
index_result = [sort_result[i][0] for i in range(5)]
print(index_result)
df['日期'][index_result]
```
