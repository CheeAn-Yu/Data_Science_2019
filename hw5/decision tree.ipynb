```python
import numpy as np
import pandas as pd
df = pd.read_csv('hw5_data.csv',sep="\s*,\s*")
df_dum = pd.get_dummies(df)
df_dum
```

```python
def gini_index(df, attribute):
    ch1_c0 = 0
    ch1_c1 = 0
    ch2_c0 = 0
    ch2_c1 = 0
    for i in range(len(df)):
        print(df.loc[i][attribute])
        if df.loc[i][attribute] == 1:
            if df.loc[i]["Class_C0"] == 1:
                ch1_c0 += 1
            elif df.loc[i]["Class_C1"] == 1:
                ch1_c1 += 1
        else:
            if df.loc[i]["Class_C0"] == 1:
                ch2_c0 += 1
            elif df.loc[i]["Class_C1"] == 1:
                ch2_c1 += 1
    ch1_num = ch1_c0+ch1_c1
    ch2_num = ch2_c0+ch2_c1
    gini_ch1 = 1 - (ch1_c0/(ch1_c0+ch1_c1))**2 - (ch1_c1/(ch1_c0+ch1_c1))**2
    gini_ch2 = 1 - (ch2_c0/(ch2_c0+ch2_c1))**2 - (ch2_c1/(ch2_c0+ch2_c1))**2
    child_gini = (ch1_num/len(df))*gini_ch1 + (ch2_num/len(df))*gini_ch2
    print(gini_ch1, gini_ch2,child_gini)
    
gini_index(df_dum, "Gender_F")
```
