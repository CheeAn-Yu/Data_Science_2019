For example:
```
p1 = Vector(1,2)
p3 = p1 + (5, 6)
p4 = p1 + [7, 8]
p5 = p1 + (7, 8, 9)

print(p3)  # You should get v(6, 8)
print(p4)  # You should get v(8, 10)
print(p5)  # you shuold also get v(8, 10)

p6 = Vector(3, 4)
n7 = p1 * p6
n8 = p1 * (5, 6)
print(n7)  # You should get 11
print(n8)  # You should get 17
```
